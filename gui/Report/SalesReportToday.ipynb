```python
import matplotlib.pyplot as plt
```

```python
from database.db_connection import create_connection
```

```python
try:
    conn=create_connection()
    cursor=conn.cursor()
    query="""
    select invoice_date,total_amount from  sales
    """
    cursor.execute(query)
    Sales=cursor.fetchall()
    print(Sales)
except Exception as err:
    print(err)
```

```python
x=[]
y=[]
for Sale in Sales:
    x.append(Sale[0])
    y.append(Sale[1])

print(x,y)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(x, y, color="purple")

plt.title("Sales Report", fontsize=14, fontweight="bold")
plt.xlabel("Date")
plt.ylabel("Revenue (â‚¹ Rupees)")

plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4, 4))
colors = [
    "#4E79A7",  # blue
    "#F28E2B",  # orange
    "#59A14F",  # green
    "#E15759",  # red
    "#B07AA1",  # purple
    "#9C755F"   # brown
]
plt.pie(
    y,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90,
    pctdistance=0.75
)

plt.legend(
    x,
    title="Date",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.title("Sales Report (Revenue Distribution)", fontsize=11, fontweight="bold")
plt.axis("equal")
plt.tight_layout()
plt.show()
```
